```python
# Data Manipulation
import numpy as np
import pandas as pd
# Plotting graphs
import matplotlib.pyplot as plt
# Machine learning
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score

# Data fetching
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
df = pdr.get_data_yahoo('AAPL', "2012-01-01", "2019-12-31")
df = df.dropna()
df = df.iloc[:,:4]
```

```python
#features
df['STD_5'] = df['Close'].rolling(window=5).mean()
df['STD_10'] = df['Close'].rolling(window=10).mean()
df['STD_20'] = df['Close'].rolling(window=20).mean()
df['STD_50'] = df['Close'].rolling(window=50).mean()
df['5_MA'] = df['Close'].shift(1).rolling(window = 5).mean()
df['10_MA'] = df['Close'].shift(1).rolling(window = 10).mean()
df['20_MA'] = df['Close'].shift(1).rolling(window = 20).mean()
df['50_MA'] = df['Close'].shift(1).rolling(window = 50).mean()
df['Upper_Band']=df['Close'].rolling(window =10).mean()+(df['STD_10']*2)
df['Lower_Band']=df['Close'].rolling(window =10).mean()-(df['STD_10']*2)
df['Exp_mov_5'] = df['Close'].ewm(span=5, adjust=False).mean().fillna(0)
df['Exp_mov_10'] = df['Close'].ewm(span=10, adjust=False).mean().fillna(0)
df['Exp_mov_20'] = df['Close'].ewm(span=20, adjust=False).mean().fillna(0)
df['Exp_mov_50'] = df['Close'].ewm(span=50, adjust=False).mean().fillna(0)
df['Correlation'] = df['Close'].rolling(window=10).corr(df['STD_10'])
df['ROC'] = ((df['Close'] - df['Close'].shift(10)) / (df['Close'].shift(10)))*100
df['per_change'] = df.Close.pct_change()
df = df.dropna()
X = df.iloc[:,:14]
X.head()
y = np.where(X['Close'].shift(-1) > X['Close'],1,0)
#X.div(X.iloc[0])
```

```python
X_train, x_validate, x_test = np.split(X.sample(frac=1), [int(.6*len(X)), int(.8*len(X))])
```

```python
y_train=np.where(X_train['Close'].shift(-1) > X_train['Close'],1,-1)
y_validate=np.where(x_validate['Close'].shift(-1) > x_validate['Close'],1,-1)
y_test=np.where(x_test['Close'].shift(-1) > x_test['Close'],1,-1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(x_test)
```

```python
#***************FIRST MODEL: LR **************************
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
predicted =log_reg.predict(X_test)
```

```python
print(metrics.confusion_matrix(y_test, predicted))
```

```python
print(metrics.classification_report(y_test, predicted))
```

```python
print(log_reg.score(x_validate,y_validate))
```

```python
print(log_reg.score(X_test,y_test))
```

```python
***********LOGISTIC REGRESSION TUNING******************
```

```python
from sklearn import  model_selection
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import TimeSeriesSplit

log_reg = LogisticRegression()
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l2']
c_values = [100, 10, 1.0, 0.1, 0.01]
# define grid search
grid = dict(solver=solvers,penalty=penalty,C=c_values)
grid_search = GridSearchCV(estimator=log_reg, param_grid=grid, n_jobs=-1, cv=3, scoring='accuracy',error_score=0)
grid_result = grid_search.fit(X_train, y_train)
```

```python
predicted_after_tuning=grid_result.predict(X_test)
```

```python
print(grid_result.score(x_validate,y_validate))
```

```python
print(grid_result.score(X_test,y_test))
```

```python
metrics.confusion_matrix(y_test,predicted_after_tuning>.5)
```

```python
***************THIRD MODEL: NEURAL NET**************************
```

```python
classifier = Sequential()
classifier.add(Dense(units = 128, kernel_initializer = 'uniform', activation = 'relu', input_dim = X.shape[1]))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss='binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, y_train, batch_size = 600, epochs = 25, verbose = 1)
```

```python
neural_net_model = classifier.predict(X_test)
_ , acc= classifier.evaluate(X_train, y_train, verbose=0)
print((acc))
```

```python
metrics.confusion_matrix(y_test, neural_net_model>.5)
```

```python
**************SECOND MODEL: RANDOM FOREST **************************
```

```python
 from sklearn.ensemble import  RandomForestClassifier
 from sklearn.model_selection import learning_curve, GridSearchCV
 from sklearn.tree import export_graphviz
 from IPython import display
```

```python
rf_model =RandomForestClassifier(n_estimators=1, max_depth=2, bootstrap=False, min_samples_leaf=1)
rf_model.fit(X_train, y_train)
rf_model.score(X_test,y_test)
```

```python
predicted_rf = rf_model.predict(X_test)
metrics.confusion_matrix(y_test,predicted_rf>.5)
```

```python

n_estimators = [10, 30, 50, 80,100]
max_depth = [3, 5, 8, 15, 25, 30]
min_samples_split = [ 5, 10, 15, 50,60]
min_samples_leaf = [1, 2, 5, 10] 

hyperF = dict(n_estimators = n_estimators, max_depth = max_depth,  
              min_samples_split = min_samples_split, 
             min_samples_leaf = min_samples_leaf)

gridF = GridSearchCV(rf_model, hyperF, cv = 3, verbose = 1, 
                      )
bestF = gridF.fit(X_train, y_train)
#skill = evaluate(bestF,y_val)
```

```python
pred_after_tuning_rf = bestF.predict(X_test)
pred_after_tuning_rf
```

```python
print("Best: %f using %s" % (bestF.best_score_, bestF.best_params_))
```
